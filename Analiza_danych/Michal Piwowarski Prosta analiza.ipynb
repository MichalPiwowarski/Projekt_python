```python
# Zajmę się teraz wartościami ostającymi. Na histogramie jest rozkład normalny, nie znajdziemy znaczących wartości 
#odstających dla 3* odchylenia standarowego, dlatego użyłem  2* odchylenie standardowe
```

```python

def find_anomalies(df):
    #define a list to accumlate anomalies
    anomalies = []
    
    # Set upper and lower limit to 2 standard deviation
    random_data_std = np.std(df['RH'])
    random_data_mean = np.mean(df['RH'])
    anomaly_cut_off = random_data_std * 2
    
    lower_limit  = random_data_mean - anomaly_cut_off 
    upper_limit = random_data_mean + anomaly_cut_off
    print(lower_limit)
    print(upper_limit)
    # Generate outliers
    for outlier in df['RH']:
        if outlier > upper_limit or outlier < lower_limit:
            anomalies.append(outlier)
        else:
            anomalies.append(0)
    return anomalies

```

```python
 outl = find_anomalies(df)
```

```python
def outliers(x):
    if x == 0:
        return False
    else:
        return True
```

```python
df['Outliers'] =  outl
```

```python
df['Outliers']=df['Outliers'].apply(lambda x: outliers(x))
```
